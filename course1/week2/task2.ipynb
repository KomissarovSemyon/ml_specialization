```python
import pandas as pd
import numpy as np
from scipy.linalg import solve
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def compute(x):
    dots = np.linspace(1, 15, 200)
    mtrx = [[i ** j for j in range(len(x))] for i in x]
    a = np.array(mtrx)
    b = f(x)
    res = solve(a, b)
    pol = np.polynomial.polynomial.Polynomial(res)
    plt.plot(dots, pol(dots))
    plt.plot(dots, f(dots))
    plt.show()
    return res
```

```python
compute(np.array([1, 15]))
compute(np.array([1, 8, 15]))
res = compute(np.array([1, 4, 10, 15]))
```

```python
np.round(res,2 )
```
